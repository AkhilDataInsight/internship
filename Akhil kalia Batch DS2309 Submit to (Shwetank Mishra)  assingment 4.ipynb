```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    # Assuming the table you want is the first one on the page
    table = soup.find('table', {'class': 'wikitable'})

    # Lists to store data
    ranks = []
    names = []
    artists = []
    upload_dates = []
    views = []

    # Loop through rows in the table
    for row in table.find_all('tr')[1:]:  # Skip the header row
        columns = row.find_all('td')
        ranks.append(columns[0].text.strip())
        names.append(columns[1].text.strip())
        artists.append(columns[2].text.strip())
        upload_dates.append(columns[3].text.strip())
        views.append(columns[4].text.strip())

    # Print or use the data as needed
    for i in range(len(ranks)):
        print(f"Rank: {ranks[i]}, Name: {names[i]}, Artist: {artists[i]}, Upload Date: {upload_dates[i]}, Views: {views[i]}")

else:
    print(f"Failed to fetch the webpage. Status code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup

# Fetch the home page to get the link to international fixtures
url_home = "https://www.bcci.tv/"
response_home = requests.get(url_home)

if response_home.status_code == 200:
    soup_home = BeautifulSoup(response_home.text, 'html.parser')

    # Find the link to international fixtures page
    fixtures_link = soup_home.find('a', {'class': 'international dropdown-item'})

    if fixtures_link:
        # Construct the URL for international fixtures
        url_fixtures = "https://www.bcci.tv" + fixtures_link['href']
        response_fixtures = requests.get(url_fixtures)

        if response_fixtures.status_code == 200:
            soup_fixtures = BeautifulSoup(response_fixtures.text, 'html.parser')

            # Assuming the details are in a specific class
            fixtures_data = soup_fixtures.find_all('div', {'class': 'fixture__info'})

            for fixture in fixtures_data:
                series = fixture.find('div', {'class': 'fixture__format-detail'}).text.strip()
                place = fixture.find('p', {'class': 'fixture__additional-info'}).text.strip()
                date = fixture.find('div', {'class': 'fixture__full-date'}).text.strip()
                time = fixture.find('span', {'class': 'fixture__time'}).text.strip()

                print(f"Series: {series}, Place: {place}, Date: {date}, Time: {time}")

        else:
            print(f"Failed to fetch international fixtures page. Status code: {response_fixtures.status_code}")

    else:
        print("Link to international fixtures not found on the home page.")

else:
    print(f"Failed to fetch the home page. Status code: {response_home.status_code}")

```

```python
import requests
from bs4 import BeautifulSoup

# Fetch the home page to get the link to the Hot 100 page
url_home = "https://www.billboard.com/"
response_home = requests.get(url_home)

if response_home.status_code == 200:
    soup_home = BeautifulSoup(response_home.text, 'html.parser')

    # Find the link to the charts page
    charts_link = soup_home.find('a', {'href': '/charts'})

    if charts_link:
        # Construct the URL for the charts page
        url_charts = "https://www.billboard.com" + charts_link['href']
        response_charts = requests.get(url_charts)

        if response_charts.status_code == 200:
            soup_charts = BeautifulSoup(response_charts.text, 'html.parser')

            # Find the link to the Hot 100 page
            hot_100_link = soup_charts.find('a', {'href': '/charts/hot-100'})

            if hot_100_link:
                # Construct the URL for the Hot 100 page
                url_hot_100 = "https://www.billboard.com" + hot_100_link['href']
                response_hot_100 = requests.get(url_hot_100)

                if response_hot_100.status_code == 200:
                    soup_hot_100 = BeautifulSoup(response_hot_100.text, 'html.parser')

                    # Assuming the details are in a specific class
                    songs = soup_hot_100.find_all('li', {'class': 'chart-list__element'})

                    # Loop through the songs
                    for song in songs:
                        song_name = song.find('span', {'class': 'chart-element__information__song'}).text.strip()
                        artist_name = song.find('span', {'class': 'chart-element__information__artist'}).text.strip()
                        last_week_rank = song.find('span', {'class': 'chart-element__meta text--center color--secondary text--last'}).text.strip()
                        peak_rank = song.find('span', {'class': 'chart-element__meta text--center color--secondary text--peak'}).text.strip()
                        weeks_on_board = song.find('span', {'class': 'chart-element__meta text--center color--secondary text--week'}).text.strip()

                        print(f"Song Name: {song_name}")
                        print(f"Artist Name: {artist_name}")
                        print(f"Last Week Rank: {last_week_rank}")
                        print(f"Peak Rank: {peak_rank}")
                        print(f"Weeks on Board: {weeks_on_board}")
                        print("------")

                else:
                    print(f"Failed to fetch Hot 100 page. Status code: {response_hot_100.status_code}")

            else:
                print("Link to the Hot 100 page not found on the charts page.")

        else:
            print(f"Failed to fetch charts page. Status code: {response_charts.status_code}")

    else:
        print("Link to the charts page not found on the home page.")

else:
    print(f"Failed to fetch the home page. Status code: {response_home.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare"

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    # Assuming the details are in a specific class or structure
    novels = soup.find_all('tr', {'class': 'datablog-row'})

    # Lists to store data
    book_names = []
    author_names = []
    volumes_sold = []
    publishers = []
    genres = []

    # Loop through the novels
    for novel in novels:
        columns = novel.find_all('td')
        book_names.append(columns[0].text.strip())
        author_names.append(columns[1].text.strip())
        volumes_sold.append(columns[2].text.strip())
        publishers.append(columns[3].text.strip())
        genres.append(columns[4].text.strip())

    # Print or use the data as needed
    for i in range(len(book_names)):
        print(f"Book Name: {book_names[i]}")
        print(f"Author Name: {author_names[i]}")
        print(f"Volumes Sold: {volumes_sold[i]}")
        print(f"Publisher: {publishers[i]}")
        print(f"Genre: {genres[i]}")
        print("------")

else:
    print(f"Failed to fetch the webpage. Status code: {response.status_code}")

```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.imdb.com/list/ls095964455/"

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    # Assuming the details are in a specific class or structure
    series_list = soup.find_all('div', {'class': 'lister-item-content'})

    # Lists to store data
    names = []
    year_spans = []
    genres = []
    run_times = []
    ratings = []
    votes = []

    # Loop through the TV series
    for series in series_list:
        name = series.find('h3', {'class': 'lister-item-header'}).a.text.strip()
        year_span = series.find('span', {'class': 'lister-item-year'}).text.strip('()')
        genre = series.find('span', {'class': 'genre'}).text.strip()
        run_time = series.find('span', {'class': 'runtime'}).text.strip()
        rating = series.find('span', {'class': 'ipl-rating-star__rating'}).text.strip()
        vote = series.find('span', {'name': 'nv'}).text.strip()

        names.append(name)
        year_spans.append(year_span)
        genres.append(genre)
        run_times.append(run_time)
        ratings.append(rating)
        votes.append(vote)

    # Print or use the data as needed
    for i in range(len(names)):
        print(f"Name: {names[i]}")
        print(f"Year Span: {year_spans[i]}")
        print(f"Genre: {genres[i]}")
        print(f"Run Time: {run_times[i]}")
        print(f"Ratings: {ratings[i]}")
        print(f"Votes: {votes[i]}")
        print("------")

else:
    print(f"Failed to fetch the webpage. Status code: {response.status_code}")

```

```python
import requests
from bs4 import BeautifulSoup

url_home = "https://archive.ics.uci.edu/"
url_show_all_datasets = "https://archive.ics.uci.edu/ml/datasets.php"

# Step 1: Navigate to the "Show All Dataset" page
response_home = requests.get(url_home)
if response_home.status_code == 200:
    soup_home = BeautifulSoup(response_home.text, 'html.parser')
    
    # Find the link to the "Show All Dataset" page
    show_all_link = soup_home.find('a', {'href': '/ml/datasets.php'})
    
    if show_all_link:
        url_show_all_datasets = "https://archive.ics.uci.edu/ml/" + show_all_link['href']

# Step 2: Fetch the content of the "Show All Dataset" page
response_show_all_datasets = requests.get(url_show_all_datasets)

if response_show_all_datasets.status_code == 200:
    soup_show_all_datasets = BeautifulSoup(response_show_all_datasets.text, 'html.parser')

    # Assuming the details are in a specific class or structure
    datasets = soup_show_all_datasets.find_all('tr', {'bgcolor': '#F1F1D4'})

    # Lists to store data
    dataset_names = []
    data_types = []
    tasks = []
    attribute_types = []
    no_of_instances = []
    no_of_attributes = []
    years = []

    # Loop through the datasets
    for dataset in datasets:
        columns = dataset.find_all('td')
        dataset_names.append(columns[0].text.strip())
        data_types.append(columns[1].text.strip())
        tasks.append(columns[2].text.strip())
        attribute_types.append(columns[3].text.strip())
        no_of_instances.append(columns[4].text.strip())
        no_of_attributes.append(columns[5].text.strip())
        years.append(columns[6].text.strip())

    # Print or use the data as needed
    for i in range(len(dataset_names)):
        print(f"Dataset Name: {dataset_names[i]}")
        print(f"Data Type: {data_types[i]}")
        print(f"Task: {tasks[i]}")
        print(f"Attribute Type: {attribute_types[i]}")
        print(f"No of Instances: {no_of_instances[i]}")
        print(f"No of Attributes: {no_of_attributes[i]}")
        print(f"Year: {years[i]}")
        print("------")

else:
    print(f"Failed to fetch the 'Show All Dataset' page. Status code: {response_show_all_datasets.status_code}")

```
